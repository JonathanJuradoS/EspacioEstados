```python
import matplotlib.pyplot as plt
import networkx as nx

# Definimos el grafo
grafo = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos y aristas
for nodo, vecinos in grafo.items():
    for vecino in vecinos:
        G.add_edge(nodo, vecino)

# Posiciones para los nodos en el gráfico
pos = nx.spring_layout(G, seed=42)

# Dibujar el grafo
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=14, arrows=True)
plt.title("Representación del Grafo para Búsqueda", fontsize=16)
plt.show()

```

```python
# DFS - Profundidad
def dfs_steps(grafo, nodo, objetivo, visitado=None, camino=None, pasos=None):
    if visitado is None:
        visitado = set()
    if camino is None:
        camino = []
    if pasos is None:
        pasos = []

    visitado.add(nodo)
    camino.append(nodo)
    pasos.append(list(camino))

    if nodo == objetivo:
        return camino, pasos

    for vecino in grafo[nodo]:
        if vecino not in visitado:
            resultado, pasos = dfs_steps(grafo, vecino, objetivo, visitado, camino.copy(), pasos)
            if resultado:
                return resultado, pasos

    return None, pasos

# BFS - Amplitud
from collections import deque

def bfs_steps(grafo, inicio, objetivo):
    visitado = set()
    cola = deque([[inicio]])
    pasos = []

    while cola:
        camino = cola.popleft()
        nodo = camino[-1]
        pasos.append(list(camino))

        if nodo == objetivo:
            return camino, pasos

        if nodo not in visitado:
            visitado.add(nodo)
            for vecino in grafo[nodo]:
                nuevo_camino = list(camino)
                nuevo_camino.append(vecino)
                cola.append(nuevo_camino)

    return None, pasos

# Ejecutamos ambos algoritmos
dfs_resultado, dfs_pasos = dfs_steps(grafo, 'A', 'F')
bfs_resultado, bfs_pasos = bfs_steps(grafo, 'A', 'F')

# Mostrar resultados
import pandas as pd
df = pd.DataFrame({'Paso DFS': pd.Series(dfs_pasos), 'Paso BFS': pd.Series(bfs_pasos)})
df

```
