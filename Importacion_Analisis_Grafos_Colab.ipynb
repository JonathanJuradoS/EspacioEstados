# 📦 Análisis de Rutas de Importación desde China hacia Perú

Este notebook analiza visualmente las tres rutas posibles para importar productos desde Shanghái hacia Lima, Perú. Se consideran los factores de costo y tiempo, usando representaciones gráficas con grafos.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear grafo general
G = nx.DiGraph()
G.add_edge('Shanghái', 'Canal de Panamá', costo='$8,000', tiempo='30 días')
G.add_edge('Canal de Panamá', 'Lima', costo='Incluido', tiempo='15 días')
G.add_edge('Shanghái', 'Cabo de Hornos', costo='$16,000', tiempo='90 días')
G.add_edge('Cabo de Hornos', 'Lima', costo='Incluido', tiempo='20 días')
G.add_edge('Shanghái', 'Los Ángeles', costo='Parte 1', tiempo='2 días')
G.add_edge('Los Ángeles', 'Panamá', costo='Parte 2', tiempo='2 días')
G.add_edge('Panamá', 'Lima', costo='$45,000', tiempo='2 días')

# Posiciones para visualización
pos = {
    'Shanghái': (0, 0),
    'Canal de Panamá': (2, 1),
    'Cabo de Hornos': (2, -1),
    'Los Ángeles': (1.5, 2),
    'Panamá': (3, 1.5),
    'Lima': (4.5, 0)
}
```

```python
# Mostrar grafo completo
plt.figure(figsize=(12, 8))
edge_labels = {(u, v): f"{d['costo']} / {d['tiempo']}" for u, v, d in G.edges(data=True)}
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("🌍 Todas las rutas posibles de importación", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Ruta 1: Marítima tradicional
plt.figure(figsize=(12, 8))
ruta = [('Shanghái', 'Canal de Panamá'), ('Canal de Panamá', 'Lima')]
colors = ['green' if (u, v) in ruta else 'gray' for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=colors, node_size=2000, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("🚢 Ruta Marítima Tradicional: Shanghái → Canal de Panamá → Lima", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Ruta 2: Marítima alternativa
plt.figure(figsize=(12, 8))
ruta = [('Shanghái', 'Cabo de Hornos'), ('Cabo de Hornos', 'Lima')]
colors = ['orange' if (u, v) in ruta else 'gray' for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=colors, node_size=2000, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("🚢 Ruta Marítima Alternativa: Shanghái → Cabo de Hornos → Lima", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Ruta 3: Aérea internacional
plt.figure(figsize=(12, 8))
ruta = [('Shanghái', 'Los Ángeles'), ('Los Ángeles', 'Panamá'), ('Panamá', 'Lima')]
colors = ['blue' if (u, v) in ruta else 'gray' for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=colors, node_size=2000, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("✈️ Ruta Aérea: Shanghái → Los Ángeles → Panamá → Lima", fontsize=14)
plt.axis('off')
plt.show()
```

## ✅ Conclusiones

- La **ruta marítima por Canal de Panamá** tiene el mejor equilibrio entre costo y tiempo (USD $8,000 / 45 días).
- La **ruta marítima alternativa vía Cabo de Hornos** es viable como contingencia pero implica alto costo y tiempo (USD $16,000 / 110 días).
- La **ruta aérea internacional** es la más rápida (6 días), pero también la más cara (USD $45,000).

Este análisis gráfico ayuda a tomar decisiones logísticas estratégicas de manera visual y comparativa.
