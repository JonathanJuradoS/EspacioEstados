# 游닍 An치lisis de Rutas de Importaci칩n desde China hacia Per칰

Este notebook analiza visualmente las tres rutas posibles para importar productos desde Shangh치i hacia Lima, Per칰. Se consideran los factores de costo y tiempo, usando representaciones gr치ficas con grafos.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear grafo general
G = nx.DiGraph()
G.add_edge('Shangh치i', 'Canal de Panam치', costo='$8,000', tiempo='30 d칤as')
G.add_edge('Canal de Panam치', 'Lima', costo='Incluido', tiempo='15 d칤as')
G.add_edge('Shangh치i', 'Cabo de Hornos', costo='$16,000', tiempo='90 d칤as')
G.add_edge('Cabo de Hornos', 'Lima', costo='Incluido', tiempo='20 d칤as')
G.add_edge('Shangh치i', 'Los 츼ngeles', costo='Parte 1', tiempo='2 d칤as')
G.add_edge('Los 츼ngeles', 'Panam치', costo='Parte 2', tiempo='2 d칤as')
G.add_edge('Panam치', 'Lima', costo='$45,000', tiempo='2 d칤as')

# Posiciones para visualizaci칩n
pos = {
    'Shangh치i': (0, 0),
    'Canal de Panam치': (2, 1),
    'Cabo de Hornos': (2, -1),
    'Los 츼ngeles': (1.5, 2),
    'Panam치': (3, 1.5),
    'Lima': (4.5, 0)
}
```

```python
# Mostrar grafo completo
plt.figure(figsize=(12, 8))
edge_labels = {(u, v): f"{d['costo']} / {d['tiempo']}" for u, v, d in G.edges(data=True)}
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("游깴 Todas las rutas posibles de importaci칩n", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Ruta 1: Mar칤tima tradicional
plt.figure(figsize=(12, 8))
ruta = [('Shangh치i', 'Canal de Panam치'), ('Canal de Panam치', 'Lima')]
colors = ['green' if (u, v) in ruta else 'gray' for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=colors, node_size=2000, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("游뚹 Ruta Mar칤tima Tradicional: Shangh치i  Canal de Panam치  Lima", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Ruta 2: Mar칤tima alternativa
plt.figure(figsize=(12, 8))
ruta = [('Shangh치i', 'Cabo de Hornos'), ('Cabo de Hornos', 'Lima')]
colors = ['orange' if (u, v) in ruta else 'gray' for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=colors, node_size=2000, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("游뚹 Ruta Mar칤tima Alternativa: Shangh치i  Cabo de Hornos  Lima", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Ruta 3: A칠rea internacional
plt.figure(figsize=(12, 8))
ruta = [('Shangh치i', 'Los 츼ngeles'), ('Los 츼ngeles', 'Panam치'), ('Panam치', 'Lima')]
colors = ['blue' if (u, v) in ruta else 'gray' for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=colors, node_size=2000, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)
plt.title("九걾잺 Ruta A칠rea: Shangh치i  Los 츼ngeles  Panam치  Lima", fontsize=14)
plt.axis('off')
plt.show()
```

## 九 Conclusiones

- La **ruta mar칤tima por Canal de Panam치** tiene el mejor equilibrio entre costo y tiempo (USD $8,000 / 45 d칤as).
- La **ruta mar칤tima alternativa v칤a Cabo de Hornos** es viable como contingencia pero implica alto costo y tiempo (USD $16,000 / 110 d칤as).
- La **ruta a칠rea internacional** es la m치s r치pida (6 d칤as), pero tambi칠n la m치s cara (USD $45,000).
